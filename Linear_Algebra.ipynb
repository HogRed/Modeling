# Introduction to Linear Algebra in Python

Linear algebra is the foundation of many machine learning algorithms and data science methods. This notebook will cover essential linear algebra concepts and demonstrate how to use Python, particularly with the NumPy library, to perform basic operations on vectors and matrices.

**Libraries Used:** We’ll primarily use `NumPy` for computations and `Matplotlib` for visualizing vectors and transformations.

Let's dive in and learn more about how a lot of ML works "under the hood"!

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Setting up default plotting configurations
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (8, 6)
```

## Vectors: The Building Blocks

Think of vectors as arrows in space. Each vector has:
- A direction (which way it points)
- A magnitude (how long it is)

In 2D space, we represent vectors with two numbers:
- First number: how far right/left (x-direction)
- Second number: how far up/down (y-direction)


```python
# Create two simple vectors
vector_a = np.array([2, 3])  # Goes 2 units right, 3 units up
vector_b = np.array([1, 4])  # Goes 1 unit right, 4 units up

print("Vector a:", vector_a)
print("Vector b:", vector_b)

# Let's see what we can do with vectors:

# 1. Vector Addition (combining vectors)
vector_sum = vector_a + vector_b
print("\nVector Addition (a + b):", vector_sum)
print("This means we:")
print("- Add the x components: 2 + 1 =", vector_sum[0])
print("- Add the y components: 3 + 4 =", vector_sum[1])

# 2. Scalar Multiplication (making vectors longer/shorter)
scalar = 2
scaled_vector = scalar * vector_a
print("\nScalar Multiplication (2 × a):", scaled_vector)
print("This means we multiply each component by 2:")
print("- X component: 2 × 2 =", scaled_vector[0])
print("- Y component: 2 × 3 =", scaled_vector[1])

# 3. Dot Product (how similar are two vectors?)
dot_product = np.dot(vector_a, vector_b)
print("\nDot Product (a · b):", dot_product)
print("This is calculated as:")
print(f"(2 × 1) + (3 × 4) = {vector_a[0]*vector_b[0]} + {vector_a[1]*vector_b[1]} = {dot_product}")
```

```python
def plot_vectors(vectors_dict):
    # Origin point
    origin = [0], [0]

    # Distinct vector colors
    colors = ['blue', 'red', 'green', 'purple']

    # Loop through vector dictionary and plot vectors based on coords
    for (name, vector), color in zip(vectors_dict.items(), colors):
        plt.quiver(*origin, vector[0], vector[1],
                   angles='xy', scale_units='xy',
                   scale=1, color=color, label=name)

    # Plot options
    plt.xlim(-1, 8)
    plt.ylim(-1, 8)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.legend()
    plt.grid(True)
    plt.title("Vector Operations Visualization")
    plt.show();

# Dictionary to hold vector info
vectors = {
    'Vector a': vector_a,
    'Vector b': vector_b,
    'a + b': vector_sum,
    '2 × a': scaled_vector
}

# Call vector plot function
plot_vectors(vectors)
```

## Matrices: Grids of Numbers

A matrix is like a spreadsheet of numbers. Each number has a specific row and column position.
We use matrices to:
- Store data in an organized way
- Transform vectors
- Solve systems of equations

```python
# Creating matrices
matrix_a = np.array([[1, 2],
                     [3, 4]])
matrix_b = np.array([[5, 6],
                     [7, 8]])

print("Matrix A:")
print(matrix_a)
print("\nThis means:")
print("- Top row: [1, 2]")
print("- Bottom row: [3, 4]")
print("\nMatrix B:")
print(matrix_b)
print("\nThis means:")
print("- Top row: [5, 6]")
print("- Bottom row: [7, 8]")

# 1. Matrix Addition
print("\n1. Matrix Addition (A + B):")
matrix_sum = matrix_a + matrix_b
print(matrix_sum)
print("We add corresponding elements:")
print("- Top left: 1 + 5 =", matrix_sum[0,0])
print("- Top right: 2 + 6 =", matrix_sum[0,1])
print("- Bottom left: 3 + 7 =", matrix_sum[1,0])
print("- Bottom right: 4 + 8 =", matrix_sum[1,1])

# 2. Scalar Multiplication
scalar = 2
print("\n2. Scalar Multiplication (2 × A):")
scaled_matrix = scalar * matrix_a
print(scaled_matrix)
print("We multiply each element by 2")

# 3. Matrix Multiplication
print("\n3. Matrix Multiplication / Dot Product:")
matrix_product = np.dot(matrix_a, matrix_b)
print(matrix_product)
print("\nThis is more complex. For each element:")
print("- Top left:", f"(1×5) + (2×7) = {1*5} + {2*7} = {matrix_product[0,0]}")
print("- Top right:", f"(1×6) + (2×8) = {1*6} + {2*8} = {matrix_product[0,1]}")
print("- Bottom left:", f"(3×5) + (4×7) = {3*5} + {4*7} = {matrix_product[1,0]}")
print("- Bottom right:", f"(3×6) + (4×8) = {3*6} + {4*8} = {matrix_product[1,1]}")
```

## Special Matrices and Properties

Some matrices have special properties that make them useful in different situations.


```python
# 1. Identity Matrix - Like the number 1 in multiplication
identity = np.identity(2)
print("Identity Matrix:")
print(identity)
print('''When you multiply any matrix by the identity matrix,
you get the original matrix back!''')

# Let's prove it:
result = np.dot(matrix_a, identity)
print("\nMatrix A × Identity Matrix:")
print(result)
print("Notice it's the same as Matrix A!")

# 2. Transpose - Flip rows and columns
transpose = matrix_a.T
print("\nTranspose of Matrix A:")
print(transpose)
print("We swapped rows and columns:")
print("- [1, 2] became [1, 3]")
print("- [3, 4] became [2, 4]")

# 3. Matrix Inverse - Like finding 1/x for a number
inverse = np.linalg.inv(matrix_a)
print("\nInverse of Matrix A:")
print(inverse)
print("\nProof that it's the inverse (should get identity matrix):")
print(np.round(np.dot(matrix_a, inverse)))  # Round to avoid FP errors
```

## Determinants

The determinant is a scalar value that can be calculated from the elements of a square matrix. It provides insights into properties like matrix invertibility. A matrix with a non-zero determinant has an inverse.

```python
# Determinant of a matrix
det_a = np.linalg.det(matrix_a)
print("Determinant of Matrix a:", det_a)
```

## Systems of Linear Equations

We can use matrices to represent systems of linear equations and solve them. For example, the system:

2x + 3y = 8  
3x + 4y = 11  

can be represented as Ax = B and solved for x.

```python
# Set up the problem
A = np.array([[2, 3], [3, 4]])  # Coefficients
B = np.array([8, 11])           # Constants

print("Our system in matrix form:")
print("A =")
print(A)
print("B =", B)

# Solve it
solution = np.linalg.solve(A, B)
print("\nSolution:")
print(f"x = {solution[0]:.2f}")
print(f"y = {solution[1]:.2f}")

# Let's verify our solution
print("\nVerifying the solution:")
print(f"2({solution[0]:.2f}) + 3({solution[1]:.2f}) = {2*solution[0] + 3*solution[1]:.2f} ≈ 8")
print(f"3({solution[0]:.2f}) + 4({solution[1]:.2f}) = {3*solution[0] + 4*solution[1]:.2f} ≈ 11")
```

## Understanding Eigenvalues and Eigenvectors

Eigenvalues and eigenvectors are fundamental concepts in linear algebra. Let's understand them correctly:

Key Concept: When a matrix transforms a vector, it usually changes both the vector's direction and magnitude. However, eigenvectors are special vectors that maintain their original direction after the transformation - they only get scaled (stretched or compressed).

Think of it this way:
- Most vectors get both rotated AND scaled when transformed by a matrix
- Eigenvectors are the special vectors that DON'T get rotated - they only get scaled
- The eigenvalue tells us the scaling factor


```python
# Create a rotation matrix (45 degrees)
theta = np.pi/4  # 45 degrees
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                           [np.sin(theta), np.cos(theta)]])

# Create a scaling matrix
scaling_matrix = np.array([[2, 0],
                          [0, 1]])

# Combine rotation and scaling (dot product)
transform_matrix = scaling_matrix @ rotation_matrix

# Create some sample vectors
regular_vector = np.array([1, 0])
theta_range = np.linspace(0, 2*np.pi, 12)
vectors = [np.array([np.cos(theta), np.sin(theta)]) for theta in theta_range]
```

```python
# Function to plot vector transformations
def plot_vector_transformations():
    # Set up figure size for clear, side-by-side visualization of transformations
    plt.figure(figsize=(15, 5))

    # Define the origin point for all quiver plots (starting point for vectors)
    origin = [0], [0]

    # 1. Plot Original Vectors (Subplot 1 of 3)
    plt.subplot(131)  # First subplot in a 1x3 grid
    for vec in vectors:
        # Plot each vector in blue with some transparency (alpha)
        plt.quiver(*origin, vec[0], vec[1], angles='xy',
                   scale_units='xy', scale=1, color='blue', alpha=0.3)

    # Set plot limits, axes, and grid for original vectors
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.axhline(0, color='black', linewidth=0.5)  # X-axis
    plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
    plt.title("Original Vectors")
    plt.grid(True)

    # 2. Plot Vectors After Rotation (Subplot 2 of 3)
    plt.subplot(132)  # Second subplot
    for vec in vectors:
        # Apply rotation matrix to each vector
        rotated = rotation_matrix @ vec

        # Plot rotated vectors in red with transparency
        plt.quiver(*origin, rotated[0], rotated[1], angles='xy',
                   scale_units='xy', scale=1, color='red', alpha=0.3)

    # Set plot limits, axes, and grid for rotated vectors
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.axhline(0, color='black', linewidth=0.5)  # X-axis
    plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
    plt.title("After Rotation (45°)")
    plt.grid(True)

    # 3. Plot Vectors After Full Transformation (Rotation + Scaling) (Subplot 3 of 3)
    plt.subplot(133)  # Third subplot
    for vec in vectors:
        # Apply the full transformation matrix (rotation + scaling) to each vector
        transformed = transform_matrix @ vec

        # Plot fully transformed vectors in green with transparency
        plt.quiver(*origin, transformed[0], transformed[1], angles='xy',
                   scale_units='xy', scale=1, color='green', alpha=0.3)

    # Set plot limits, axes, and grid for transformed vectors
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.axhline(0, color='black', linewidth=0.5)  # X-axis
    plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
    plt.title("After Rotation + Scaling")
    plt.grid(True)

    # Ensure layout fits well within figure boundaries
    plt.tight_layout()
    plt.show();

# Call function to execute and visualize vector transformations
plot_vector_transformations()
```

Now let's find and visualize the eigenvectors of a transformation matrix. Eigenvectors represent directions in which the transformation acts by simply scaling, without changing direction. This means applying the matrix to an eigenvector results in a vector that is either stretched or compressed along the same line.

We’ll calculate the eigenvalues and eigenvectors of our matrix, then visualize these transformations to understand how they are scaled by their corresponding eigenvalues.

```python
# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

def plot_eigenvector_transformation():
    plt.figure(figsize=(12, 5))

    # Origin for all vectors
    origin = [0], [0]

    # Plot original vectors (unit circle vectors)
    for theta in np.linspace(0, 2 * np.pi, 12):
        v = np.array([np.cos(theta), np.sin(theta)])
        plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1,
                   color='blue', alpha=0.2)

    # Define distinct colors for each eigenvector and its transformation
    colors = ['red', 'orange']

    # Highlight eigenvectors and their transformations with unique colors
    for i, (eigval, eigvec) in enumerate(zip(eigenvalues, eigenvectors.T)):
        # Original eigenvector
        plt.quiver(*origin, eigvec[0], eigvec[1], angles='xy', scale_units='xy',
                   scale=1, color=colors[i],
                   label=f'Eigenvector {i+1}')

        # Transformed eigenvector
        transformed = A @ eigvec
        plt.quiver(*origin, transformed[0], transformed[1], angles='xy',
                   scale_units='xy', scale=1, color=colors[i], alpha=0.5,
                   label=f'Transformed Eigenvector {i+1}')

    # Set limits for clarity
    plt.xlim(-6, 3)
    plt.ylim(-6, 3)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title("Eigenvectors and Their Transformations")
    plt.legend(loc="upper right")
    plt.grid(True)
    plt.show()

# Plot eigenvector transformations
plot_eigenvector_transformation()

# Output eigenvalues and eigenvectors
print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```


## Applications of Eigenvalues / Eigenvectors
1. Principal Component Analysis (PCA):
   - Eigenvectors represent the main directions of variation in data
   - Eigenvalues tell us how much variation occurs in each direction

2. Stress Analysis in Engineering:
   - Eigenvectors show principal stress directions
   - Eigenvalues give the magnitude of stress

3. Quantum Mechanics:
   - Eigenvectors represent stable states
   - Eigenvalues represent observable measurements
